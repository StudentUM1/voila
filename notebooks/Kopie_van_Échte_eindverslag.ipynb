```python
# import all the necessary things
import pandas as pd

import altair as alt
!pip install iso3166
from iso3166 import countries as ct

import ipywidgets as ipy
import numpy as np
```

```python
# import dataset
df = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
# Check if all the countries in the dataset are actual countries
countries_iso = [c.name for c in ct]
countries_df  = list(pd.unique(df["location"]))
countries = [c for c in countries_iso if c in countries_df]
```

```python
# Use some descriptive statistics
```

```python
# To be able to use all dates, I will have to change all NaN to 0
# The reason for this is that it is necessary to have at least a number to be used in the graphs
df = df.replace(np.nan, 0)
```

```python
# let the user choose what they want to analyse
Chosen_subject = ipy.Dropdown(
    options=[('New cases', 'new_cases'), ('New deaths','new_deaths')],
    description='Subject:',
    disabled=False,
)
```

```python
# Let the user choose the color used in the graph
Chosen_colour = ipy.Dropdown(
    options=['Blue', 'Cyan', 'Green', 'Black', 'Magenta', 'Red', 'Yellow', 'Purple'],
    description='Color:',
    disabled=False,
)
```

```python
# Let the user choose the first country they want to analyse
Chosen_country_1 = ipy.Dropdown(
    options=countries,
    description='Country 1:',
    disabled=False,
)
```

```python
# Let the user choose the second country they want to analyse
Chosen_country_2 = ipy.Dropdown(
    options=countries,
    description='Country 2:',
    disabled=False,
)
```

```python
# how does the mean during a certain month compare from one country to the other
print("Your research question:")
print('How do the means ', Chosen_subject.value, ('compare between'))
```

```python
# Put all the information from the user in a logical place
ipy.HBox([Chosen_subject, Chosen_colour, Chosen_country_1, Chosen_country_2])
```

```python
# Get the right information for country 1 to be used in the graph
df2_1 = df[['location', 'date', Chosen_subject.value]]

df3_1 = df2_1.loc[(df2_1['location'] == Chosen_country_1.value)]
```

```python
# Get the right information for country 2 to be used in the graph
df2_2 = df[['location', 'date', Chosen_subject.value]]

df3_2 = df2_2.loc[(df2_2['location'] == Chosen_country_2.value)]
```

```python
#create the necessary data to make the first graph
#The reason for using iloc is because the graph above doesnt work with what I wanted to do
#To be very honest, I don't know how all of this works, but somehow it does
df_new_1 = df3_1.iloc[:, [0,1,2]]

df_new_1['date'] = pd.to_datetime(df['date'])

df4_1 = df_new_1.set_index('date')

df4_1_month = df4_1.resample('M').mean()

df5_1 = df4_1_month.reset_index()
```

```python
#create the necessary data to make the second graph
df_new_2 = df3_2.iloc[:, [0,1,2]]

df_new_2['date'] = pd.to_datetime(df['date'])

df4_2 = df_new_2.set_index('date')

df4_2_month = df4_2.resample('M').mean()

df5_2 = df4_2_month.reset_index()
```

```python
# Show the chart for the first country
print('Here you can see the graph for', Chosen_country_1.value)
fig1 = alt.Chart(df5_1).mark_line(opacity=0.7, color=Chosen_colour.value).encode(
    x='date',
    y=Chosen_subject.value,
)
fig1
```

```python
print('Here you can see the graph for', Chosen_country_2.value)
fig2 = alt.Chart(df5_2).mark_line(opacity=0.7, color=Chosen_colour.value).encode(
    x='date',
    y=Chosen_subject.value,
)
fig2
```

```python
print('As a conclusion, you can see above how', Chosen_country_1.value, 'compares to', Chosen_country_2.value)
```
